```python
%matplotlib inline
import sympy as sy
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
# the data used in this notebook is get from here: https://www.kaggle.com/datasets/rounakbanik/the-movies-dataset

ratings = pd.read_csv('ratings_small.csv', dtype={'movieId': 'object', 'userId': 'object'})
movies = pd.read_csv('movies_metadata.csv')
df = ratings.merge(movies, left_on='movieId', right_on='id')[['original_title',  'userId', 'rating']]
df = df.rename(columns={'original_title': 'movie'})
```

```python
most_active_users = df.userId.value_counts()[:10].index
```

```python
users_by_movie = df[(df['userId'].isin(most_active_users))].groupby('movie').count()['userId'] 
most_common_movies = users_by_movie[users_by_movie <= 9].sort_values(ascending=False).index[:15]
```

```python
selected_samples = df[(df['userId'].isin(most_active_users)) & (df['movie'].isin(most_common_movies))]
```

```python
ratings_df = selected_samples.pivot(values='rating', columns='userId', index='movie')
ratings_df
```

```python
ratings_df.to_csv('small_movie_ratings.csv')
```
