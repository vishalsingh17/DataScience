# Texthero

## Clean text

```python
import numpy as np
import pandas as pd
import texthero as hero

df = pd.DataFrame(
    {
        "text": [
            "Today is a beautiful   day",
            "There are 3 ducks in this pond",
            "This is. very   cool.",
            "This won't work",
            np.nan,
        ]
    }
)

df.text.pipe(hero.clean)
```

```python
from texthero import preprocessing

custome_pipe = [
    preprocessing.remove_brackets,
    preprocessing.remove_stopwords,
    preprocessing.remove_whitespace,
]
```

```python
df = pd.DataFrame(
    {
        "text": [
            "Today is a beautiful   day",
            "[square] or (round) brackets are not fun",
        ]
    }
)

hero.clean(df["text"], custome_pipe)
```

```python
data = pd.read_csv("processed_articles.csv")
data
```

```python
data["clean_abstract"] = data["ABSTRACT"].pipe(hero.clean)
```

```python
data["pca"] = data["clean_abstract"].pipe(hero.tfidf).pipe(hero.pca)
```

```python
data['pca'].head(10)
```

```python
plot = hero.scatterplot(data, col="pca", color="Topic", title="Research Papers")
```
