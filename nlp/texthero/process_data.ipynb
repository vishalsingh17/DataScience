```python
import pandas as pd

papers = pd.read_csv("articles.csv")
papers
```

```python
topics = papers.select_dtypes("int").drop(columns="ID").columns
```

```python
papers["total_topics"] = papers[topics].sum(axis=1)
papers = papers[papers.total_topics == 1]
```

```python
transformed_papers = papers.melt(
    id_vars=["TITLE", "ABSTRACT"], value_vars=topics, var_name="Topic"
)
data = transformed_papers[transformed_papers.value == 1].drop(columns="value").sample(5000)
data.head(10)
```

```python
data.to_csv("processed_articles.csv", index=False)
```
